```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#CARGAMOS UN DATASET
data = sns.load_dataset('iris')
data, set(data['species'])

```

```python
#GRAFICO DE DISPERCION
#hue es para diferenciar por colores cada especie
sns.scatterplot(x='sepal_length', y='sepal_width', data=data, hue='species')
plt.xlabel('Longitud del sépalo')
plt.ylabel('Ancho del sépalo')
plt.title('Gráfico de dispersión del iris')
plt.legend(title='Especie')
plt.show()
```

```python
#GENERAR UNA GRAFICA DE RIDGEPLOT
#ESTA GRAFICA MUESTRA LA DISTRIBUCION DE UNA VARIABLE PARA CADA CATEGORIA
setosa = data[data['species'] == 'setosa']
versicolor = data[data['species'] == 'versicolor']
virginica = data[data['species'] == 'virginica']

#CONFIGURAMOS LA FIGURA Y LOS EJES
fig, ax = plt.subplots(figsize=(9, 6))
plt.xlabel('Longitud del pétalo')

#CREAMOS LA GRAFICA DE RIDGEPLOT
sns.kdeplot(data = setosa["petal_length"], label='Setosa', ax=ax, fill=True)
sns.kdeplot(data = versicolor["petal_length"], label='Versicolor', ax=ax, fill=True)
sns.kdeplot(data = virginica["petal_length"], label='Virginica', ax=ax, fill=True)

#AJUSTAMOS LA POSICION DE LA LEYENDA
ax.legend(loc='upper right')
plt.title('Gráfico de ridgplot del iris')
plt.show()
```

```python
# GRFICA DE DISEPRCION 
sns.set_style('whitegrid')
pallete = sns.color_palette('husl', 3)

sns.scatterplot(x='sepal_length', 
                y='sepal_width', 
                data=data, hue='species', 
                palette=pallete)

plt.xlabel('Longitud del sépalo')
plt.ylabel('Anchura del sépalo')
plt.title('Gráfico de dispersión del iris')

plt.show()
```

```python
#EJEMPLO DE GRAFICO AVANZADO
#PROBLEMA: VISUALIZAR EL PRECIO DE LOS AUTOS Y VER SU DEVALUACION CON EL TIEMPO

#ESTABLECEMOS LA SEMILLA DE LOS NIMEROS ALEATORIOS PARA REPRODUCIR LOS RESULTADOS
np.random.seed(42)

edad_autos = np.random.randint(0, 20, size=100)
precio_autos = 30 - edad_autos + np.random.normal(-3, 3, size=100)

data = pd.DataFrame({
    'Edad': edad_autos,
    'Precio': precio_autos
})
data
```

```python

#DENTRO DEL SUBPLOT LOS NUMEROS DENTRO DE LOS PARENTESIS Y FUERA DEL FITSIN SON LAS CUANTAS FILAS Y COLUMNAS QUE QUEREMOS
fig, ax = plt.subplots(2,2, figsize=(12,10))

#CREAR EL GRAFICO DE DISPERCION
#EL ALPHA ES LA TRANSPARENCIA DE LOS PUNTOS, ENTRE MAS BAJO SEA EL ALPHA MAS TRANSPARENTE SERA EL GRAFICO
ax[0,0].scatter(data["Edad"], data["Precio"], alpha=0.5)
ax[0,0].set_xlabel('Edad del auto')
ax[0,0].set_ylabel('Precio del auto')
ax[0,0].set_title('Gráfico de dispersión')

#CREAMOS HISTOGRAMA
#kde GRAFICA DE LA DISTRIBUCION DE LOS DATOS, SI ES TRUE SE DIBUJA LA CURVA DE LA DISTRIBUCION
sns.histplot(data['Edad'], ax=ax[0,1], kde=True,  color='orange', bins=10,)
ax[0,1].set_xlabel('Edad del auto')
ax[0,1].set_ylabel('Frecuencia')
ax[0,1].set_title('Histograma de la edad de los autos') 

#HISTOGRAMA DE LA VARIABLE PRECIO
sns.histplot(data["Precio"], ax = ax[1,0], kde=True, color='green', bins=10)
ax[1,0].set_xlabel('Precio del auto')
ax[1,0].set_ylabel('Frecuencia')
ax[1,0].set_title('Histograma del precio de los autos')
```
