```python
import pandas as pd 
```

```python
#DICCIONARIO
data = {
    "Nombre" : ["Juan", "Luis", "Ana", "Maria"],
    "Edad" : [22,25,23,30],
    "Ciudad" : ["MAchala", "Quito", "Cuenca", "Guayaquil"]
}

data, type(data)
```

```python
#GENERAMOS UN DATAFRAME A PARTIR DEL DICCIONARIO

df = pd.DataFrame(data=data)
df
```

```python
#SELECCIONAMOS UNA COLUMNA
nombres = df["Nombre"]
print(nombres, type(nombres))
```

```python
#SELECCIONAMOS MAS DE UNA COLUMNA
df[["Nombre", "Ciudad"]]
```

```python
#FILTRAR POR INIDCE
fila = df.loc[2]
print(fila)
```

```python
#FILTRAR POR CONDICION
df[df["Edad"] > 23]
```

```python
#DOBLE CONDICION
#df[(df["Edad"] > 23) & (df["Nombre"].str.startswith("M"))]
filtro = (df["Edad"] > 23) & (df["Nombre"].str.startswith("M"))
df[filtro]
```

```python
#filtrar por query
df.query("Edad > 23 and Nombre.str.startswith('M')")
```

```python
df[df["Nombre"].isin(["Luis", "Ana", "Carlos"])]
```

```python
# EXPORTAR UN DATADRAME A CSV
df.to_csv("datos.csv")
```

```python
#IMPORTAR UN CSV A UN DATAFRAME
#index_col=0 para que la primera columna se convierta en el indice del dataframe
import_df = pd.read_csv("datos.csv", index_col=0)
import_df
```
