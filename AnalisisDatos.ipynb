```python
#AGRUPACION DE DATOS
data = {
    "Nombre": ["Juan", "Ana", "Luis", "Laura", "Pedro", "Carla"],
    "Ciudad": ["Madrid", "Barcelona", "Madrid", "Valencia", "Barcelona", "Madrid"],
    "Edad": [25, 33,30,28,45,38],
    "Puntuacion": [80, 90 ,85, 88, 75, 91]
    }
df = pd.DataFrame(data)
df
```

```python
#AGRUPAR DATOS POR CIUDAD
grouped = df.groupby("Ciudad")
print(grouped.groups)
```

```python
#CACULAR LA SUMA DE LAS EDADES Y PUNTUACIONES POR CIUDAD
agregated_data = grouped.agg(
    {
        "Edad": "mean",
        "Puntuacion": "sum"
    }
)
agregated_data
```

```python
#DEFINIMOS UNA FUNCION DE AGREGACION PERSONALIZADA
#EL RANGO EN ESTE CASO ES AL VARIACION ENTRE LE MAXIMO VALOR Y EL MINIMO
def rango(series):
    return series.max() - series.min()

#APLICAMOS LA FUNCION DE AGREGACION PERSONALIZADA A LA COLUMNA DE EDAD
agregated_data_custom = grouped.agg(
    {  
        "Edad": rango,
        "Puntuacion": rango
    }
    )
print (agregated_data_custom)
```

```python
#AGRUPACION DE DATOS
data["Categoria"] = ["A", "B", "A", "B", "A", "B"]
df = pd.DataFrame(data)
df
```

```python
#AGREGAR UNA NUEVA COLUMNA CON UNA FUNCION LAMBDAAGRUPACION DE DATOS POR CIUDAD Y CATEGORIA
grouped_mult = df.groupby(["Ciudad", "Categoria"])
grouped_mult.groups
```

```python
#AGREGADO DE CALCULAR LAS UMA DE EDADES Y PUNTUACIONES POR CIUDAD Y CATEGORIA
agregated_data_multi = grouped_mult.agg(
    {
        "Edad": "sum",
        "Puntuacion": "mean"
    }
)
agregated_data_multi
```

```python
#AGREGAMOS UNA COLUMAN
data = {
    "Nombre": ["Juan", "Ana", "Luis", "Laura"],
    "Edad" : [25,33,30,28]
}
df = pd.DataFrame(data)
df
```

```python
#AQUI AGREGAMOS UNA COLUMA SIN MODIFICAR EL DATASET ORIGINAL
df["Ciudad"] = ["Madrid", "Barcelona", "Madrid", "Valencia"]
df
```

```python
#GENERAMOS UNA NUEVA FILA
new_row= pd.Series({"Nombre": "Pedro", "Edad": 45, "Ciudad": "Bilbao"})

#AGREGAMOS LA FILA
# EL .T ES PARA HACER LA TRANSPUESTA YA QUE ESTE SE CRE ORIGINALMENTE COMO UNA COLUMNA Y NOSOTROS ESTAMOS AGREGANDO UNA FILA
#ignore_index=True PARA QUE SE REASIGNE EL INDICE DEL DATAFRAME DESPUES DE AGREGAR LA FILA
df = pd.concat([df, new_row.to_frame().T], ignore_index=True)
df
```
