```python
import numpy as np
```

```python
data = {
    "Nombre" : ["Juan", "Luis", "Ana", "Maria", "Carlos"],
    "Edad" : [22,25,np.nan,30,28],
    "Ciudad" : ["MAchala", "Quito", "Cuenca", None, "Ambato"]
}
df = pd.DataFrame(data)
df 
```

```python
#RELLENAR LOS VALORES FALTANTES
#Metodo fillna() para rellenar los valores faltantes
#METODO mean DEVUELVE LA MEDIA DE LA COLUMNA
df_fill = df.fillna(
    {
    "Edad" : df["Edad"].mean(),
    "Ciudad" : "Desconocida"
    }
)
df_fill
```

```python
#ELIMINAMOS LAS FILAS CON VALORES FALTANTES
df_drop = df.dropna()
df_drop
```

```python
#REEMPLAZAR VALORES ESPECIFICOS DE UNA COLUMNA
#AQUI SE PUEDEN REEMPLAZAR VARIOS VALORES DE UNA COLUMNA O FILA A LA VEZ
def_reem = df.replace(
    {
        "Ciudad" : {None : "Desconocida"}
    }
)
def_reem
```

```python
#INTERPOLACION DE VALORES FALTANTES
df_interploado = df.copy()
df_interploado["Edad"] = df_interploado["Edad"].interpolate()
df_interploado
```

```python
#TRATAR VALORES DUPLICADOS
data_dupl = {
    "Nombre" : ["Ana", "Juan", "Pedro", "Maria", "Luis", "Ana", "Juan"],
    "Edad" : [22,25,None,23,20,22,25],
    "Ciudad" : ["Barcelona", "Madrid", "Valencia", None, "Bilbao", "Barcelona", "Madrid"]
}
df_dupl = pd.DataFrame(data_dupl)
df_dupl
```

```python
#ELIMINAR FILAS DUPLICADAS, TIENEN QUE SER EXACTAMENTE IGUALES PARA SER CONSIDERADAS DUPLICADAS 
df_sin_duplicados = df_dupl.drop_duplicates()
df_sin_duplicados
```

```python
#PARA REENOMBRAR LAS COLUMNAS
df_renombrado = df.rename(columns={"Nombre":"Name", "Edad":"Age", "Ciudad":"City"})
df_renombrado
```

```python
#ORDENAR POR UNA COLUMNA
columnas_ordenadas = ["Ciudad", "Edad", "Nombre"]
df_ordenado = df[columnas_ordenadas]
df_ordenado
```

```python
#TRANSFORMACION DE DATOS
def cuadrado(x):
  return x**2

#SE AGREGA UNA NUEVA COLUMNA CON ESTA FUNCION 
df["Edad_Cuadrado"] = df["Edad"].apply(cuadrado)
df
```
