```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#PCA SE USA PARA REDUCIR LA DIMENSIONALIDAD DE LOS DATOS, MANTENIENDO LA MAYOR VARIANZA POSIBLE. 
#ES UNA TÉCNICA DE ANÁLISIS DE DATOS QUE TRANSFORMA UN CONJUNTO DE VARIABLES CORRELACIONADAS EN UN CONJUNTO DE VARIABLES NO CORRELACIONADAS LLAMADAS COMPONENTES PRINCIPALES.
from sklearn.decomposition import PCA
```

```python
df = sns.load_dataset('iris')
#CON EL ILOC SELECCIONAMOS TODAS LAS FILAS Y TODAS LAS COLUMNAS MENOS LA ÚLTIMA, QUE ES LA COLUMNA DE CLASES.
df.iloc[:, :-1]
```

```python
pca_model = PCA(n_components=2)
df_pca = pca_model.fit_transform(df.iloc[:, :-1])
df_pca
```

```python
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=df['species'].astype('category').cat.codes)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()
```

```python
#TRANSFORMACION LOGARITMICA DE VARIABLE

sns.displot(data=df, x='sepal_length',hue='species', kde=True)
plt.show()
```

```python
#Tranformacion logaritmica de variable
df['sepal_length_log'] = np.log(df['sepal_length'])
sns.displot(data=df, x='sepal_length_log',hue='species', kde=True)
plt.show()
```

```python
#ESCALADO Y NORMALIZACION DE DATOS
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
std_scaler = StandardScaler()
df_std_scaler = std_scaler.fit_transform(df.iloc[:, :-2])
df_std_scaler
```
