```python
import matplotlib.pyplot as plt
import numpy as np  
import math
```

```python
#EJEMPLO DE UNA LINEA SIMPLE
x = [0,1,2,3,4,5]
y = [0,1,2,3,4,5]
plt.plot(x,y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Grafico de Linea Simple')
plt.legend(['Linea'])
plt.scatter(x,y)
plt.show()
```

```python
#GRAFICO DE BARRAS
categorias = ['A', 'B', 'C', 'D', 'E']  
valores = [10, 15, 7, 12, 20]

plt.bar(categorias, valores)

plt.title('Grafico de Barras')
plt.xlabel('Categorias')    
plt.ylabel('Valores')

plt.gca().set_axisbelow(True)  # hace que el grid quede detrás
plt.grid(True)

plt.show()
```

```python
#PERSONALIAZACION DE ESTILOS
plt.style.available
plt.style.use('ggplot') 
```

```python
x = [0,1,2,3,4,5]
y = [0,1,4,9,16,25]

plt.plot(x,y, 
         color='blue',
         linestyle='--',
         marker='o', 
         markersize=8, 
         markerfacecolor='red'
         )

plt.title('Grafico de Linea Personalizado')
plt.ylabel("Valores medidos")
plt.xlabel("Tiempo")
plt.legend(['Linea'])
plt.show()


```
