```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style="whitegrid")

df = sns.load_dataset("tips")
#head me imprime las primeras filas del dataframe, para tener una idea de como es la estructura de los datos
df.head()
```

```python
#ANALISIS UNIVARIADO
#EXAMINA SOLO UNA VARABLE A LA VEZ CON GRAFICO DE CAJAS, HISTOGRAMAS. ETC
#NOS DEJA VER LA VARIACION Y COMPORTAMIENTO DE CADA VARIABLE POR SEPARADO

sns.histplot(data=df, x="total_bill", kde=True)
plt.plot()
```

```python
sns.boxplot(data=df, x="total_bill")
plt.plot()
```

```python
sns.kdeplot(data=df, x="total_bill", fill=True)
plt.plot()
```

```python
#ANALISIS BIVARIADO
#EXAMINA LA RELACION ENTRE DOS VARIABLES CON GRAFICOS DE DISPERSION, DI

sns.scatterplot(data=df, x="total_bill", y="tip")
plt.plot()
```

```python
sns.barplot(data=df, x="day", y="total_bill")
plt.xlabel("Día de la semana")
plt.ylabel("Cuenta total")
plt.plot()
```

```python
sns.barplot(
    data=df,
    x="sex",
    y="tip",
    hue="sex",
    palette=["skyblue", "lightcoral"],
    legend=False,
    dodge=False
)

plt.xlabel("Sexo del cliente")
plt.ylabel("Propina")
plt.show()
```

```python
sns.pairplot(data = df, hue= "sex")
plt.show()
```

```python
#GRAFICO DE BARRAS PROPINAS PROMEIDO POR SEXO
sns.barplot(data=df, 
            x="sex", 
            y="tip", 
            hue="sex", 
            errorbar=None, 
            palette="pastel", 
            legend=False,
            dodge=False)
plt.xlabel("Sexo del cliente")
plt.ylabel("Propina promedio")
plt.title("Propina promedio por sexo")
plt.show()
```

```python
sns.barplot(data=df,
            x="day", 
            y="tip",
            hue="day",
            dodge=False,
            errorbar=None,
            palette="pastel",)

plt.xlabel("Día de la semana")
plt.ylabel("Propina promedio")
plt.title("Propina promedio por día de la semana")
plt.show()
```

```python
#GARFICO DE BARRAS DE PROPINA PROMEDIO POR DIA DE LA SEMANA Y SEXO
sns.barplot(
    data=df,
    x="day",
    y="tip",
    hue="sex",
    #dodge=False,
    errorbar=None,
    #palette="pastel",
)

plt.xlabel("Día de la semana")
plt.ylabel("Propina promedio")
plt.title("Propina promedio por día de la semana y sexo")
plt.show()
```

```python
#PROPINAS PROMEDIO POR SEXO Y TIEMPO
sns.barplot(
    data=df,
    x="time",
    y="tip",
    hue="sex",
    #dodge=False,
    errorbar=None,
    #palette="pastel",
)

plt.xlabel("Tiempo")
plt.ylabel("Propina promedio")
plt.title("Propina promedio por tiempo y sexo")
plt.show()
```

```python
#GRAFCIO DE DISPERSION ENTRE TAMAÑO DE LA CUENTA Y LA PROPINAS
sns.scatterplot(
    data=df,
    x="size",
    y="tip"
)

plt.title("Relación entre tamaño de la cuenta y propina")
plt.xlabel("Tamaño de la cuenta")   
plt.ylabel("Propina")
plt.show()
```

```python
# GRAFICO DE BARRAS DE PROPINAS PROMEDIO POR TAMAÑO DE LA FIESTA
sns.barplot(
    data=df,
    x="size",
    y="tip",
    errorbar=None,
    hue = "size",
    palette="pastel",
)

plt.title("Propina promedio por tamaño de la fiesta")
plt.xlabel("Tamaño de la fiesta")   
plt.ylabel("Propina promedio")
plt.show()
```

```python
#CALCULAMOS EL PORCENTAJE DE PROPINA Y LO AGREGAMOS AL DATAFRAME

df["tip_percentage"] = (df["tip"] / df["total_bill"]) * 100
df
```

```python
sns.barplot(
    data=df,
    x="sex",
    y="tip_percentage",
    errorbar=None,
    hue = "sex",
    dodge=False,
    palette="pastel",
)

plt.xlabel("Sexo del cliente")
plt.ylabel("Porcentaje de propina")
plt.title("Porcentaje de propina por sexo")
plt.show()
```
